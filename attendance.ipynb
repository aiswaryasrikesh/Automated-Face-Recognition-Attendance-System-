```python
import numpy as np
import os
import cv2
import face_recognition as fr
from datetime import datetime
```

```python
path='picture'
```

```python
mylist=os.listdir(path)
mylist
```

```python
imgs=[]
classnames=[]
for i in mylist:
    imgpath=os.path.join(path,i)
    current_img=cv2.imread(imgpath)
    imgs.append(current_img)
    classnames.append(i.split('.')[0])
```

```python
classnames
```

```python
def faceencodings(picture):
    encodelist=[]
    for img in imgs:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frames=fr.face_locations(img)
        face_encode=fr.face_encodings(img,face_in_frames)[0]
        encodelist.append(face_encode)
    return encodelist
    
```

```python
encodelist_known_face=faceencodings(imgs)
print(encodelist_known_face)
```

```python

marked = []
if os.path.exists("attendance.csv"):
    with open("attendance.csv", "r") as f:
        for line in f:
            name = line.split(",")[0]
            if name not in marked:
                marked.append(name)
video = cv2.VideoCapture(0)
while True:
    success, frame = video.read()
    if not success:
        break

    frame1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face_in_frame = fr.face_locations(frame1)
    face_encoder = fr.face_encodings(frame1, face_in_frame)

    for encoder, faceloc in zip(face_encoder, face_in_frame):
        match = fr.compare_faces(encodelist_known_face, encoder)
        distance = fr.face_distance(encodelist_known_face, encoder)
        index = np.argmin(distance)
        if match[index]:
            name = classnames[index]
            if name not in marked:
                marked.append(name)
                now = datetime.now()
                date = now.strftime("%Y-%m-%d")
                time = now.strftime("%H-%M-%S")
                with open("attendance.csv", "a") as f:
                    f.write(f"{name},{date},{time}\n")
        else:
            name = "unknown"
    cv2.imshow("frame", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
video.release()
cv2.destroyAllWindows()
```
